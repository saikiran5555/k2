Characteristics: Manhattan distance is less sensitive to outliers compared to Euclidean distance because it calculates the distance by traversing along the coordinate axes. It's often used when the movement is restricted to orthogonal (axis-aligned) paths, such as grid-based environments.
How Might This Difference Affect KNN Performance?
Sensitivity to Scale: Euclidean distance is sensitive to differences in scale between features, as it considers the square of the differences. In contrast, Manhattan distance considers absolute differences, making it less sensitive to scale variations.
Effect of Outliers: Due to its sensitivity to the magnitude of differences, Euclidean distance can be more affected by outliers than Manhattan distance, which sums the absolute differences. Thus, Manhattan distance might be more robust in the presence of outliers.
Data Geometry: The choice between Euclidean and Manhattan distance can significantly affect the notion of "closeness" between points, which in turn impacts the performance of KNN. For datasets with irregular or nonlinear structures, Euclidean distance might capture relationships better, while Manhattan distance might be more appropriate for datasets with a grid-like or structured geometry.
Computational Complexity: Manhattan distance can be computationally cheaper to compute compared to Euclidean distance, especially in higher-dimensional spaces, due to its simple absolute difference calculation.
